```python
import bs4
import urllib.request as url
```

```python
httpRes = url.urlopen('http://stats.espncricinfo.com/ci/content/records/283792.html')
```

```python
page = bs4.BeautifulSoup(httpRes)
```

```python
table = page.find('table',class_='engineTable')
```

```python
print(table.text)
```

```python
data_1 = table.text.split("\n\n")
```

```python
del data_1[0]
```

```python
for i in range(len(data_1)):
    data_1[i] = data_1[i].split("\n")
```

```python
for i in range(len(data_1)):
    del data_1[i][0]
```

```python
del data_1[1]
```
